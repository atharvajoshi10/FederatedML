```python
from datasets import load_dataset
```

# Loading the Data



```python
train_dataset = load_dataset("glue", "mrpc", split ="train").shuffle(42)
validation_dataset = load_dataset("glue", "mrpc", split ="validation").shuffle(42)
```

# Splitting the data into 8 parts to simulate federated learning

```python
for i in range(8):
  train_dataset.shard(num_shards=8, index = i, contiguous=False) \
               .save_to_disk("/content/drive/MyDrive/FML Data/Federated Split/train/{}/train.hf".format(i))

  validation_dataset.shard(num_shards=8, index = i, contiguous=False).save_to_disk("/content/drive/MyDrive/FML Data/Federated Split/validation/{}/validation.hf".format(i))
```
